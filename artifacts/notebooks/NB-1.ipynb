```python
import pandas as pd
```

```python
df = pd.read_csv('bank_marketing_data.csv')
df.head()
```

```python
df["term_deposit_subscribed"].value_counts()
```

```python
df_no = df[df["term_deposit_subscribed"] == "no"]
df_yes = df[df["term_deposit_subscribed"] == "yes"]
```

```python
df_no_sample = df_no.sample(len(df_yes))
len(df_no_sample)
```

```python
new_df = pd.concat([df_yes, df_no_sample])
print(len(new_df))
```

```python
new_df.head()
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(new_df, random_state=24, test_size=0.2, stratify=new_df["term_deposit_subscribed"])

print(len(train_data))
print(len(test_data))
```

```python
train_data.to_csv("artifacts/data/train_data_balanced.csv", index=False)
test_data.to_csv("artifacts/data/test_data_balanced.csv", index=False)
```
